Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python

nasa_coordinate = [28.562302, -80.577356]  # 根據第一個發射場的座標
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# 添加 MarkerCluster 到地圖
marker_cluster = MarkerCluster().add_to(site_map)

# 定義成功和失敗的顏色
outcome_colors = {
    1: "green",  # 成功
    0: "red"     # 失敗
}

# 為每個發射場創建並添加 Marker 和 Circle，並將其添加到 MarkerCluster 中
for index, row in spacex_df.iterrows():
    # 根據 Class 設置顏色
    color = outcome_colors[row["class"]]
    
    # 添加圓形標記到 MarkerCluster
    folium.Circle(
        location=[row["Lat"], row["Long"]],
        radius=100000,  # 半徑可依需求調整
        color=color,
        fill=True,
        popup=folium.Popup(f"{row['Launch Site']} - {'Success' if row['class'] == 1 else 'Failed'}", parse_html=True)
    ).add_to(marker_cluster)

    # 添加標記到 MarkerCluster
    folium.Marker(
        location=[row["Lat"], row["Long"]],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % row["Launch Site"],
        )
    ).add_to(marker_cluster)

# 添加赤道線
equator_coordinates = [[0, -180], [0, 180]]  # 赤道的線 (緯度為 0)
folium.PolyLine(
    locations=equator_coordinates,
    color="red",
    weight=2.5,  # 線的寬度
    opacity=1
).add_to(site_map)

# 顯示地圖
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# 初始化地圖
nasa_coordinate = [28.562302, -80.577356]  # 根據第一個發射場的座標
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# 添加 MarkerCluster 到地圖
marker_cluster = MarkerCluster().add_to(site_map)

# 定義成功和失敗的顏色
outcome_colors = {
    1: "green",  # 成功
    0: "red"     # 失敗
}

# 為每個發射場創建並添加 Marker 和 Circle，並將其添加到 MarkerCluster 中
for index, row in spacex_df.iterrows():
    # 根據 Class 設置顏色
    color = outcome_colors[row["class"]]
    
    # 添加圓形標記到 MarkerCluster
    folium.Circle(
        location=[row["Lat"], row["Long"]],
        radius=100000,  # 半徑可依需求調整
        color=color,
        fill=True,
        popup=folium.Popup(f"{row['Launch Site']} - {'Success' if row['class'] == 1 else 'Failed'}", parse_html=True)
    ).add_to(marker_cluster)

    # 添加標記到 MarkerCluster
    folium.Marker(
        location=[row["Lat"], row["Long"]],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % row["Launch Site"],
        )
    ).add_to(marker_cluster)

# 添加赤道線
equator_coordinates = [[0, -180], [0, 180]]  # 赤道的線 (緯度為 0)
folium.PolyLine(
    locations=equator_coordinates,
    color="red",
    weight=2.5,  # 線的寬度
    opacity=1
).add_to(site_map)

# 添加 MousePosition 插件來顯示滑鼠位置的經緯度
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# 假設你使用 MousePosition 標記下來的海岸線的座標
coastline_lat = 28.563  # 替換為實際海岸線座標
coastline_lon = -80.570  # 替換為實際海岸線座標

# 選擇其中一個發射場 (例如 CCAFS LC-40) 計算與海岸線的距離
launch_site_lat = spacex_df.loc[0, "Lat"]
launch_site_lon = spacex_df.loc[0, "Long"]

# 計算發射場與海岸線的距離
distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# 在地圖上標記最近的海岸線點，並顯示距離
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
    )
).add_to(site_map)

# 創建發射場和海岸線之間的 PolyLine
coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]

# 添加 PolyLine 到地圖
lines = folium.PolyLine(locations=coordinates, weight=1, color="blue")
site_map.add_child(lines)

# 顯示地圖
site_map
```
